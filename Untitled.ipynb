```python
import plotly.express as px
import pandas as pd
import numpy as np
from pathlib import Path
from pprint import pprint
from matplotlib import pyplot as plt
```

```python
resString = Path("./hudData/prepilecap updates/footCxCResults.csv")
tabResString = Path("./hudData/prepilecap updates/footCxCTabResults.csv")
results = pd.read_csv(resString);
tabResults = pd.read_csv(tabResString);
results.head()
```

```python
def getMaxMin(column):
    Max = abs(column).max()
    Min = -Max
    return Max, Min
    
```

```python
xStr = "Fx"
yStr = "Fz"
df_orig = results.copy()
```

```python
# %matplotlib inline
df = df_orig[df_orig["set"].str.contains('ULS B')]
df2 = df[df["Location"].str.contains('MC')]


colors = ["b","g","r","c","m","y","k"]
fig = [0]*6
ax = [0]*6

xMax,xMin = getMaxMin(df2[xStr])
yMax,yMin = getMaxMin(df2[yStr])
2
j=0
for key,group in df2.groupby("Location"):
    fig,ax= plt.subplots(figsize=(8,6))
    plt.xlim(1.05*xMax,1.05*xMin)
    plt.ylim(1.05*yMax,1.05*yMin)
    plt.title(key)
    j+=1
    i=0
    grouped = group.groupby('set')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x=xStr, y=yStr, label=key, color=colors[i])
        i+=1
```
